```python
import psycopg2
import uuid
import random
from datetime import datetime
"""
The idea is to gen data for a "product_categoies" table that hold data repensent which category a product belong to
"""


# Connect to PostgreSQL database
conn = psycopg2.connect(
    dbname='test_nam_1',
    user='postgres',
    password='admin',
    host='localhost',
    port='5432'
)

cursor = conn.cursor()


```

```python
# Fetch product IDs
cursor.execute("SELECT id FROM public.products;")
product_ids = [row[0] for row in cursor.fetchall()]

# Fetch category IDs
cursor.execute("SELECT id FROM public.categories;")
category_ids = [row[0] for row in cursor.fetchall()]

print(product_ids)


print("\n\n\n\n\n")
print(len(category_ids))
```

```python
# Generate insert statements
insert_statements = []
with open("product_categories.sql", "w") as file:
    for product_id in product_ids:
        # Randomly select a category ID
        category_id = random.choice(category_ids)
        
        # Generate a new UUID for the product_category record ID
        record_id = uuid.uuid4()
        
        # Define timestamps
        created_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        updated_at = created_at
        
        insert_statement = (
            f"INSERT INTO product_categories (id, category_id, product_id, created_at, updated_at, deleted_at) "
            f"VALUES ('{record_id}', '{category_id}', '{product_id}', '{created_at}', '{updated_at}', NULL);\n"
        )
        file.write(insert_statement)

# Clean up
cursor.close()
conn.close()
```
